```python
Student_Data = {'Student_id': ['AAA-123-2211','BBB-542-8542','CCC-854-8541','AAA-856-7458','DDD-745-7841'],
                'First_Name': ['Sam','Jim','Rick','Ben','Kevin'],
                'Last_Name': ['Edison','Mon','Dany','Denil','Pocket']}

student= pd.DataFrame (Student_Data, columns = ['Student_id','First_Name','Last_Name'])
student
```

```python
Student_Course= {'stu_id':['AAA-123-2211','BBB-542-8542','CCC-854-7458','AAA-856-7458',
                               'DDD-745-7841'],'course':['ML','DS','RPA','BI','DL'],
                 'duration':[3,4,2,5,2]}
course = pd.DataFrame(Student_Course)
course
```

```python
# concatenating or combining two DataFrames along a particular axis


print (pd.concat([student,course]))
```

```python
print (pd.concat([student,course],axis=1))
```

```python
# Merging two or more DataFrames based on a common column or index
# It's similer to SQL Joins



print (pd.merge(student,course,left_on='Student_id',right_on='stu_id', how='right'))
```

```python
print (pd.merge(student,course,left_on='Student_id',right_on='stu_id', how='left'))
```

```python
pd.merge(student,course,left_on= 'Student_id',right_on ='stu_id', how = 'inner')
```

```python
pd.merge(student,course,left_on= 'Student_id',right_on ='stu_id', how = 'outer')
```
