# Histogram

```python
age = [4,5,7,10,12,45,32,42,22,21,28,35,14,16,20,33,33,35,38,32,42,45,48,18,19,26]
age_group= [0,10,20,30,40,50]

rcParams['figure.figsize']=10,6

plt.hist(age,age_group,histtype='bar', rwidth=.8,color='y',label='Age Group')

plt.title ('Graph Three')
plt.xlabel ('X- Axis')
plt.ylabel ('Y-Axis')
plt.legend()
plt.show()
```

```python
age=[10,7,8,5,12,45,32,42,22,21,28,29,35,14,16,20,33,33,35,38,32,42,45]
male_age=[0,10,20,30,40,50]
female_age=[0,5,10,15,20,25,30]
plt.hist(age,male_age,histtype='bar',rwidth=0.50,color='g',label='Male')
plt.hist(age,female_age,histtype='bar',rwidth=0.50,color='r',label='female')

plt.title('Age Measurment')
plt.xlabel ('X- Axis')
plt.ylabel ('Y-Axis')
plt.legend()
plt.show()
```

Multi Dount Chart

```python
sleep = [7,6,4,5,8]
eat = [2,3,4,2,1]
work =[8,9,10,8,9]
play=[4,5,3,3,2]
plt.figure(figsize= (10,10))

plt.pie(sleep,pctdistance=0.92,autopct='%1.1f%%, \n (sleep)',radius=1.0)
plt.pie(eat,pctdistance=0.88,autopct='%1.1f%%, \n (eat)',radius=0.80)
plt.pie(work,pctdistance=0.86,autopct='%1.1f%%, \n (work)',radius=0.60)
plt.pie(play,pctdistance=0.84,autopct='%1.1f%%, \n (play)',radius=0.40)
circle=plt.Circle((0,0),0.20, fc='white')
plt.gca().add_artist(circle)
plt.title('Mut. Dount Chart')
plt.show()
```

# Multiple Graph

```python
x1=np.arange(10,30,2)
y1 = [1,5,6,7,8,3,4,6,5,4]

x2= np.arange (10,30,2)
y2 = [5,4,2,8,4,6,3,4,5,4]

plt.figure(figsize=(15,11))
plt.subplot(221)
plt.plot(x1,y1)
plt.title('Graph One')

plt.subplot(222)
plt.plot(x2,y2)
plt.title('Graph Two')

plt.show()
```

```python
def f(t):
    return (np.exp(-t)*np.cos(2*np.pi*t))

t1=np.arange(10,20,2)
t2= np.arange (11,21,2)

plt.figure(figsize=(15,8))

plt.subplot (221)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.title ('Graph 1')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')


plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.title ('Graph 2')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')

plt.show()
```

```python
t2= np.arange (11,21,2)

plt.figure(figsize=(15,8))

plt.subplot (221)
plt.plot(t2,(t2),'bo',t2,(t2))
plt.title ('Graph 1')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')

```
