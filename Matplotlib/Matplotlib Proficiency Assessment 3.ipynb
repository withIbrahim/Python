```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pylab import rcParams
```

```python
x=[2,4,6,8,10]
y=[1,2,1.5,5,3]

rcParams['figure.figsize']=10,6

plt.plot(x,y,'r--o',alpha=0.8)
plt.show()
```

```python
x =[1,2,3,4,5]
y= [15,10,12,18,20]
x1= [11,13,15,17,19]
y1= [20,17,25,12,28]

plt.plot(x,y, 'r--*',label= 'Days',)
plt.plot(x,x1,'g:.',label='Max.T',)
plt.plot(x,y1,'c-o',label='Mic.T')

plt.legend(loc='upper left',fontsize='medium',shadow=True)
plt.grid(alpha=0.5)

plt.title('Weater')
plt.xlabel('Days')
plt.ylabel('Tempreture')


plt.show()
```

```python
company =['Google','Amazon','Microsoft','Facebook']
revenue=[90,136,89,27]

plt.bar(company,revenue, color=['b','r','g','c'])
plt.title('US Tech Stocks')
plt.xlabel('Company')
plt.ylabel('Revenue')

plt.show()
```
